### 03- How to Rename Columns ?


```python
df = pd.DataFrame({'A': [1, 2, 3, 4],'B': [5, 6, 7, 8],'C': [9, 10, 11, 12]})
df
```

```python
df.rename(columns={'A': 'a-col', 'B': 'b-col', 'C': 'c-col'}, inplace=True)
df
```

```python
df.columns=['a', 'b', 'c']
df
```

```python
# to replace any character , strings or numbers in a column
df.columns=df.columns.str.replace('-', '_')
df
```

```python
#adding prefix to columns
df=df.add_prefix('sk_')
df
```

```python
#adding suffix to columns
df=df.add_suffix('_sk')
df
```
